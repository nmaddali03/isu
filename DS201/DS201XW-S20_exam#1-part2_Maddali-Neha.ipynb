```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Automobile_price_data_clean-f18.csv')
df1 = df.groupby(df['make']).price.mean()
df1.plot.bar(color=['brown', 'orchid', 'red', 'pink', 'orange', 'darkcyan', 'green', 'purple', 'blue','lightgreen', 'gray'])
plt.title('Average of Price by Make')
# Question 1)
plt.show()
```

```python
toyotaPrice = df.loc[df['make']=='toyota', ['price']]
toyotaGreater = toyotaPrice.loc[toyotaPrice['price'] > 8000]
toyotaGreater.count()
# Question 2) 22 toyota cars are priced above USD 8,000
```

```python
hatchback = df.loc[df['body-style'] == 'hatchback', ['city-mpg']]
hatchback.mean()
# Question 3) 26.602941 mpg is the average city-mpg for hatchback cars
```

```python
df.boxplot(column = 'price', by = 'body-style', figsize = (15,10), fontsize = 12, rot = 45)
plt.title('Price Boxplot grouped by Body-style')
# Question 4)
plt.show()
```

```python
df.boxplot(column = 'price', by = 'make', figsize = (15,10), fontsize = 12, rot = 90)
plt.title('Price Boxplot grouped by make')
# Question 5)
plt.show()
```
