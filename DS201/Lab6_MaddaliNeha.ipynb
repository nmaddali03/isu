```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('drinks.csv')
```

```python
df.head()
```

```python
df.loc[(df['continent'] == 'Europe') & (df['beer_servings'] > 50)]
```

```python
df.loc[(df['continent'] == 'Europe') & (df['beer_servings'] > 50), ['country']]
```

```python
df2 = df.groupby(['continent']).size()
```

```python
df2.plot.bar()
plt.show()
```

```python
df3 = df.groupby("continent").beer_servings.sum()
```

```python
df3.plot.bar()
plt.show()
```

```python
df4 = df.groupby("continent").beer_servings.mean()
```

```python
df4
```

```python
df4.plot.bar()
plt.show()
```

```python
df5 = df.boxplot(column = 'beer_servings',by = 'continent', rot=45, figsize=(10,7), fontsize=10, grid=False)
plt.show()
```
